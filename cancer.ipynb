```python
A=[]
df.drop('id',axis=1,inplace=True)
df.drop('Unnamed: 32',axis=1,inplace=True)
df.drop('smoothness_mean',axis=1,inplace=True)
# size of the dataframe
len(df)
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
df.head()
```

```python

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

standardized_data = scaler.fit_transform(df)

print(standardized_data)
```

```python
y=df["diagnosis"]
# X
# print(X)
# Y.values.reshape(-1,1)
# print(X)
X=df.drop("diagnosis",axis=1)
# x
# y.values.reshape(-1,1)
# print(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
# X.reshape(1, -1)
preds=knn.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
print("correct prediction",accuracy_score(y_test,preds))
print("wrong prediction",1-accuracy_score(y_test,preds))
cm=confusion_matrix(y_test,preds)
print(cm)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, preds)
A.append(accuracy)
print('Accuracy:',accuracy)
print('Precision: %.3f' % precision_score(y_test, preds))
print('Recall: %.3f' % recall_score(y_test, preds))
print('F1 Score: %.3f' % f1_score(y_test, preds))
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression(normalize=True)
lin_reg.fit(X_train,y_train)
y_pred2= lin_reg.predict(X_test)  
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score

def cross_val(model):
    pred = cross_val_score(model, X, y, cv=10)
    return pred.mean()

def print_evaluate(true, predicted):  
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)
    print('__________________________________')
    
def evaluate(true, predicted):
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    return mae, mse, rmse, r2_square
```

```python
test_pred = lin_reg.predict(X_test)
train_pred = lin_reg.predict(X_train)

print('Test set evaluation:\n_____________________________________')
print_evaluate(y_test, test_pred)
print('Train set evaluation:\n_____________________________________')
print_evaluate(y_train, train_pred)

results_df = pd.DataFrame(data=[["Linear Regression", *evaluate(y_test, test_pred) , cross_val(LinearRegression())]], 
                          columns=['Model', 'MAE', 'MSE', 'RMSE', 'R2 Square', "Cross Validation"])
```
